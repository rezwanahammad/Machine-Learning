### Here review,education is ordinal encoding, gender,purchased is nominal encoding

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']]) #this is optional but giving it by yourself is good
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train # here Good and PG becomes 2 and the rest follows this order
```
